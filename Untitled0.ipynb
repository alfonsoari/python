```
import numpy as np
import matplotlib.pyplot as plt
```

```
ntl = lambda x: 10*np.log10(x)
ltn = lambda x: 10**(x/10)

def cos(frekuensi,tipe,tower,user,d):
    if frekuensi in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frekuensi in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frekuensi/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frekuensi))+18.33*(np.log10(frekuensi))-40.94
    else :
        Cm='tipe salah'
    
    auser = (1.1*np.log10(frekuensi)-0.7)*user-(1.56*np.log10(frekuensi)-1.8)
    Lp = Ca+(Cb*np.log10(frekuensi))-(13.83*np.log(tower))-auser+((44.9-6.55*np.log10(tower))*np.log10(d))-Cm
    
    return Lp
```

```
tower = 80
user = 1.5
dmin=300 
dmax=1000 
N = 200
ptr = 30
gtr = 10
gre = 2
bw = 5
d=np.random.randint(dmin,dmax,N)
```

```
#fungsi daya terima, snr, kapasitas
def pterima(ptr,pl,gtr,gre):
  ptr1 = ntl(ptr)
  daya = ptr - pl + gtr + gre
  daya1 = ltn(daya)
  return daya1

def snr(pterima,bw):
  noise = -174 + 10*np.log10(bw)
  noise1 = ltn(noise)
  snr = pterima/noise1
  return snr

def ks(snr,bw):
  ks = bw*2*np.log(1+snr)
  return ks
```

```
#Daya Terima
#daya terima f sama
def pterimaurban():
  f=1800 
  loc='urban'
  pl=cos(f,loc,tower,user,d)
  pterimafix = pterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=pterimafix[np.argsort(d)]
  return d_sort,Urban_sort

def pterimaopen():
  f=1800 
  loc='open'
  pl=cos(f,loc,tower,user,d)
  pterimafix = pterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=pterimafix[np.argsort(d)]
  return d_sort,Urban_sort

def pterimasuburban():
  f=1800 
  loc='suburban'
  pl=cos(f,loc,tower,user,d)
  pterimafix = pterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=pterimafix[np.argsort(d)]
  return d_sort,Urban_sort
  
d11,pl11 = pterimaurban()
d12,pl12 = pterimaopen()
d13,pl13 = pterimasuburban()

#daya terima lokasi sama
def pterima700():
  f=700
  loc='urban'
  pl=cos(f,loc,tower,user,d)
  pterimafix = pterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=pterimafix[np.argsort(d)]
  return d_sort,Urban_sort

def pterima900():
  f=900
  loc='urban'
  pl=cos(f,loc,tower,user,d)
  pterimafix = pterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=pterimafix[np.argsort(d)]
  return d_sort,Urban_sort

def pterima1800():
  f=1800 
  loc='urban'
  pl=cos(f,loc,tower,user,d)
  pterimafix = pterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=pterimafix[np.argsort(d)]
  return d_sort,Urban_sort

d21,pl21 = pterima700()
d22,pl22 = pterima900()
d23,pl23 = pterima1800()
```

```
plt.figure()
plt.plot(d11,pl11,'m', label='urban')
plt.plot(d12,pl12,'c', label='open')
plt.plot(d13,pl13,'y', label='suburban')
plt.legend()
plt.title('Daya Terima')
plt.xlabel('Jarak (m)')
plt.ylabel('Daya Terima (miliwatt)')
plt.show()

plt.figure()
plt.plot(d21,pl21,'m', label='700 Mhz')
plt.plot(d22,pl22,'c', label='900 Mhz')
plt.plot(d23,pl23,'y', label='1800 Mhz')
plt.legend()
plt.title('Daya Terima')
plt.xlabel('Jarak (m)')
plt.ylabel('Daya Terima (miliwatt)')
plt.show()
```
